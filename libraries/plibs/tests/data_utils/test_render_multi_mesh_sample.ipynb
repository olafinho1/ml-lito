```python
import os.path

%load_ext autoreload
%autoreload 2

# we assume plibs is outside blender_rendering/ and in the same root_dir
import os
import sys

repo_root = "../.."
sys.path.insert(0, repo_root)  # repo root
sys.path.insert(0, "../../..")  # for blender_rendering at same level as repo root

import data_utils
import structures
```

```python
mesh_filenames = [
    "../../../blender_rendering/assets/0a0f1b107acb4b94a8211e11ab69a67f.glb",
    "../../../blender_rendering/assets/boom_karts_rocket_vehicle.glb",
]
mesh_filenames = [os.path.abspath(f) for f in mesh_filenames]
```

```python
out_dir = "tmp_output"
out_dict = data_utils.render_multi_mesh_sample(
    out_dir=out_dir,
    mesh_filenames=mesh_filenames,
    num_regular_images=4,
    regular_width_px=512,
    regular_height_px=512,
    num_random_images=3,
    random_width_px=256,
    random_height_px=256,
    overwrite=True,
    seed=None,
    read_result_to_rgbd=True,
)
```

```python
rgbd: structures.RGBDImage = out_dict["rgbd_regular"]
rgbd.camera.save(
    output_dir=os.path.join(out_dir, "regular_camera"),
    overwrite=True,
    save_ply=True,
    save_individual_ply=True,
    world_frame_size=1,
    camera_frame_size=0.3,
)
```
